```python
import numpy as np
import gym
import scipy
```

```python
env = gym.make('Taxi-v2')
env.reset()
env.render()
# it's an taxi game
# +20 for correct dropoff
# -10 for incorrect pickup or drop
# -1 for every timestep
# taxi will turn green when the passenger is abroad
# R G Y B are location
```

```python
# random search algorithm
total_count = 0
for _ in range(1000):
    reward=0
    count=0
    while reward != 20:
        state,reward,done,info = env.step(env.action_space.sample())
        # env.render()
        count +=1
    total_count += count
print("Average number of steps : {}".format(total_count/float(1000)))
```

```python
# q-learning algorithm
q = np.zeros([env.observation_space.n,env.action_space.n])
average_count = 0
gamma=0.6 #discount factor

for episode in range(1000):
    done = False
    count = 0
    state = env.reset()
    average_count=0
    
    while not done :
        action = np.argmax(q[state])
        next_state,reward,done,info = env.step(action)
        # env.render()
        count+=1
        q[state,action]= reward + gamma*(np.max(q[next_state]))
        state=next_state
    
    average_count += count 
# print(q)
    print("Average number of steps : {}".format(average_count))
# print("Average number of steps : {}".format(average_count/float(1000)))
```
